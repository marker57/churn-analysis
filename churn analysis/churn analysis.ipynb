```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Reading Datasets

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
# df.reset_index(drop=True)
df.head()
```

```python
df["SeniorCitizen"]=df["SeniorCitizen"].replace(0,'No')
df["SeniorCitizen"]=df["SeniorCitizen"].replace(1,'Yes')
df.head()
```

```python
df.isnull().sum()
# there is no missing value
```

```python
df.shape
```

```python
df.info
```

```python
df.describe()
```

```python
# how many total missing values do we have?
permiss=df.isnull().sum()
total_cells = np.product(df.shape)
total_missing = permiss.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

## visualization

```python
df['gender'].value_counts()
```

```python
labels=['Male','Female']
gen=df['gender'].value_counts()
fig,ax=plt.subplots()
plt.pie(gen,labels=labels,explode=[0,0.1],autopct='%1.2f%%')
plt.title('No of Gender with curn',fontsize=(32))
plt.legend()
plt.show()
```

```python
gh=df['gender'].value_counts().plot(kind='bar',figsize=(15,7),color=sns.color_palette('dark'),rot=0)
for p in gh.patches:
    gh.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
plt.xlabel('Gender',color='red',weight='bold')
plt.show()

```

## The average monthly charge

```python
avgcharge=df['MonthlyCharges'].mean()
print('The average monthly charge is:',avgcharge)
```

```python
avgmale=df[df['gender']=='Male']['MonthlyCharges'].mean()
avgfemale=df[df['gender']=='Female']['MonthlyCharges'].mean()
print('The average monthly charges for male is:',avgmale)
print('The average monthly charge for Female is: ',avgfemale)
```
